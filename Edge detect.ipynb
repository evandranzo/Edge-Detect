This notebook demonstrates a simple use of the Canny edge detector.

```python
import cv2 as cv
import matplotlib.pyplot as plt
```

```python
# Upload and put your image filename here
img = cv.imread("ross.jpg")
  
# Setting parameter values for Canny edge detector
t_lower = 70  # Lower Threshold 
t_upper = 130  # Upper threshold 
  
edge = cv.Canny(img, t_lower, t_upper) 

img = cv.cvtColor(img, cv.COLOR_BGR2RGB)
plt.imshow(img)
plt.show()

plt.imshow(edge)
plt.show()

# The first attempt is an atristic canny
```
